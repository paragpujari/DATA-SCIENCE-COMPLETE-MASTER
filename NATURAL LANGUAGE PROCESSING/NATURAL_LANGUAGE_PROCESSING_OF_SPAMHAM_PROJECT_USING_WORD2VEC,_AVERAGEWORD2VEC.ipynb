```python
# import the library for lemmatizer
from nltk.stem import WordNetLemmatizer

# call the object for lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('wordnet')
from nltk.corpus import stopwords
```

```python
# get the corpus text  for all the words except the special characters
messages = data.head(1000)
print(messages)

corpus = []
for i in range(0, len(messages)):
  # remove all the special characters from the text
  review = re.sub('[^a-zA-Z]',' ',messages['messages'][i])
  # convert the text into lower case
  review = review.lower()
  # split the text into words
  review = review.split()

  # stem the words to the root form
  review = [lemmatizer.lemmatize(word) for word in review]
  # join all the including the blanks
  review = ' '.join(review)
  # append all the words into the empty corpus
  corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, messages['messages']) if i<1]
```

```python
print(corpus)
```

```python
# import the library for sent_tokenize and simple process

from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
# consider all the sentences in the corpus
for sentences in corpus:
  # convert the corpus text into sentences
  sent_tokens = sent_tokenize(sentences)
  # now convert the sentences into lower case
  for i in sent_tokens:
    words.append(simple_preprocess(i))
```

```python
# training the model with Word2Vec

model = gensim.models.Word2Vec(words)
```

```python
# Perform Average Word2Vec Operation

def average_word_2_vec(doc):
  return(np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key], axis = 0))
```

```python
from tqdm import tqdm
```

```python
# apply averageword2vec for the entire sentences

X = []
for i in tqdm(range(len(words))):
  X.append(average_word_2_vec(words[i]))
```
