```python
# Filter 
#print(dataset)
x = dataset['hp'] > 200
results = dataset[x]
print(results)
```

```python
#print(dataset)
res = dataset['vs'] > 0
ans = dataset[res]
print(ans)
```

```python
# filter all the records from the dataset whose mpg is greater than 20
x = dataset['mpg'] > 20.0
ans = dataset[x]
print(ans)
```

```python
# filter all the records in which the MPG is greater than 10 and mpg is less than 20
x = ((dataset['mpg'] > 10) & (dataset['mpg'] < 20))
res = dataset[x]
#print(res)

# sort them in ascending order
res.sort_values(by = 'mpg', ascending = True)
```

```python
# Apply lambda function so as to find cyl * 2
f = lambda x : x*2
res['cyl'] = res['cyl'].apply(f)
print(res)
```

```python
# Apply lambda function so as to find disp / 100
f = lambda x : x / 10
res['disp'] = res['disp']. apply(f)
print(res)
```
