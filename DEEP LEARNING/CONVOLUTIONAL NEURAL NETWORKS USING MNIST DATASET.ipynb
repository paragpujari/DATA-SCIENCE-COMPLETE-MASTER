# Importing all the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.keras import datasets
import keras
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
```

# Laoding the dataset

```python
(X_train, Y_train) , (X_test, Y_test) = datasets.mnist.load_data()
```

# Shape of the datasets

```python
print(X_train.shape)
print(X_test.shape)


print(Y_train.shape)
print(Y_test.shape)
```

# Reshaping the training and testing data for the input

```python
# training data
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], X_train.shape[2], 1))

# test data
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], X_test.shape[2], 1))
```

```python
print(X_train[:5])
```

```python
print(X_test[:5])
```

# DATA PREPROCESSING

```python
X_train = X_train.astype('float')

X_test  = X_test.astype('float')
```

```python
X_train = X_train / 255.0

X_test  = X_test / 255.0
```

```python
print(X_train[:5])
```

```python
print(X_test[:5])
```

# PLOTTING THE FEW SETS OF MNIST DATASETS

```python
# Depicting the size of the figure
fig = plt.figure(figsize = (18,5))

for i in range(15):
    # adding the sub plot in the figure
    ax = fig.add_subplot(2,10,i+1,xticks = [], yticks = [])
    # depicting the gray scale figure of the hand written dataset
    ax.imshow(np.squeeze(X_train[i]), cmap = 'gray')
    # depicting the title of the handwritten dataset
    ax.set_title(Y_train[i])
```

# DEPICTING THE SHAPE OF INPUT IMAGE

```python
# shape of training data
print(X_train.shape)
```

```python
# shape of the input image
input_shape = X_train.shape[1:]
print(input_shape)
```

# BUILD AN ANN MODEL TO PERFORM THE MNIST HANDWRITTEN DATASET

```python
ann = Sequential([
    # Arrange all the input layer in vector form
    Flatten(input_shape = (28, 28, 1)),
    # Add the first input layer
    Dense(3000, activation = 'relu'),
    # Add the second input layer
    Dense(1000, activation = 'relu'),
    # Add the output layer
    Dense(10,  activation = 'softmax')
])
```

# COMPILE THE ANN MODEL

```python
ann.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

# TRAIN THE ANN MODEL

```python
ann.fit(X_train, Y_train, epochs = 10)
```

# EVALUATE THE ANN MODEL

```python
ann.evaluate(X_test, Y_test)
```

# SUMMARY OF ANN MODEL

```python
ann.summary()
```

# LOSS AND ACCURACY OF ANN MODEL

```python
loss, accuracy = ann.evaluate(X_test, Y_test, verbose = 0)
print("Loss of the model is:", loss)
print("Accuracy of the model is:", (accuracy * 100.0))
```

# CLASSIFICATION REPORT AND CONFUSION MATRIX OF ANN MODEL

```python
from sklearn.metrics import classification_report, confusion_matrix

# find the predicted value
pred_y = ann.predict(X_test)

# find out the predicted class of output
pred_class_y = [np.argmax(i) for i in pred_y]
print(pred_class_y)
```

```python
# To get the classification report
classification_report(Y_test, pred_class_y)
```

```python
# To get the confusion matrix
confusion_matrix(Y_test, pred_class_y)
```

# BUILD THE CNN MODEL TO PERFORM THE MNIST HANDWRITTEN DATASET

```python
# build the cnn model to perform the handwritten dataset
cnn = Sequential([
    # To add the first convolutional layer
    Conv2D(filters = 32, kernel_size = (3,3), activation = 'relu', input_shape = (28,28,1)),
    # Add the Max Pooling layer 
    MaxPooling2D((2,2)),
    
    # To add the second convolutional layer
    Conv2D(filters = 32, kernel_size = (3,3), activation = 'relu'),
    # Add the Max Pooling layer
    MaxPooling2D((2,2)),
    
    
    # Perform the Drop out layer on Max Pooled layer to reduce the number of neurons
    Dropout(0.5),
    
    # Perform the flattening layer to bring all the neurons in vector form
    Flatten(),
    
    # Add the dense layer to the CNN Model
    Dense(500, activation = 'relu'),
    
    # Add the dense layer to the CNN Model
    Dense(10, activation = 'softmax')
])
```

# COMPILE THE CNN MODEL

```python
cnn.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

# TRAIN THE CNN MODEL

```python
cnn.fit(X_train, Y_train, epochs = 10, batch_size = 128, verbose = 2, validation_split = 0.1)
```

# SUMMARY OF CNN MODEL

```python
cnn.summary()
```

# EVALUATE THE CNN MODEL

```python
cnn.evaluate(X_test, Y_test)
```

# LOSS AND ACCURACY OF CNN MODEL

```python
loss, accuracy = cnn.evaluate(X_test, Y_test)
print("Loss of the model is:", loss)
print("Accuracy of the model is:", (accuracy * 100.0))
```

# CLASSIFICATION REPORT AND CONFUSION MATRIX OF CNN MODEL

```python
from sklearn.metrics import classification_report, confusion_matrix

# find the predicted value
pred_y = cnn.predict(X_test)

# find out the predicted class of output
pred_class_y = [np.argmax(i) for i in pred_y]
print(pred_class_y)
```

```python
# To find out the classification report
classification_report(Y_test, pred_class_y)
```

```python
# To find out the confusion matrix
cm = confusion_matrix(Y_test, pred_class_y)
```

```python
# To represent the confusion matrix in heatmap
import seaborn as sns
sns.heatmap(cm, annot = True)
```
