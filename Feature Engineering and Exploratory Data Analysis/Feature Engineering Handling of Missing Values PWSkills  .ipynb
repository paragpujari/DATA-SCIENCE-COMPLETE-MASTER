# Imputaion Missing Technique

## 1. Mean Value Imputation

It refers to the process of replacing all the null values with the mean of the particular column.

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.isnull().sum()
```

```python
# fill all the NULL Values of Age with the mean value of the ages
df['age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['age_mean','age']]
```

# This method works well when we have normally distributed data.

## 2. Median Imputation Technique

## It is used when we have outliers in the dataset.

```python
# fill all the NULL Values of the ages with the median value of the age 
df['age_median'] = df['age'].fillna(df['age'].median())
```

```python
df[['age_median', 'age_mean', 'age']]
```

# 3. Mode Imputation Technique

It works well with categorical values

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
x = df['embarked'].mode()[0]
```

```python
# replace all the null values in the embarked column with the mode value
df['embarked_mode'] = df['embarked'].fillna(x)
```

```python
df[['embarked_mode', 'embarked']]
```

```python
df['embarked_mode'].isna().sum()
```
