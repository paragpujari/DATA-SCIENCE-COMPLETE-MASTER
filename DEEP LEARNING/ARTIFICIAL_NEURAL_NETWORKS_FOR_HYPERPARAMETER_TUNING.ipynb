# Import all the necessary libraries for the neural networks

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# keras = It can be defined as the framework which has all the neural network libraries and it helps in running the neural networks
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Loading the dataset
data = pd.read_csv('/content/diabetes.csv')
print(data)
```

```python
# first five rows of the dataset
print(data.head())
```

```python
# divide the dataset into independent and dependent variables
predictors = data.iloc[:,0:8].values
print(predictors)

response = data.iloc[:,8].values
print(response)
```

# The main objective of the Artificial Neural Network is to predict whether the person has diabetis or not based on the diagnsotic measurements



# Use of the countplot to show the summary statistics of the number of people who is diabetic and who is non-diabetic

```python
import seaborn as sns
# use of the count plot for depicting the diabetic number and  non-diabetic number
data['Outcome'].value_counts().plot(kind = 'bar')
```

```python
# splitting the datas into training and testing data
X_train, X_test, y_train, y_test = train_test_split(predictors, response, test_size = 0.2, random_state = 0)
```

# Training the Artificial Neural Model

```python
# initialize the sequential model
kerasmodel = Sequential()
# Add the input layer to the initialized Sequential model
kerasmodel.add(Dense(12, input_dim = 8, activation = 'relu'))
# Add the dropout funtion to the input layer
kerasmodel.add(Dropout(0.5))
# Add the second hidden layer
kerasmodel.add(Dense(8, activation = 'relu'))
# Add the dropout function to the second hidden layer
kerasmodel.add(Dropout(0.5))
# Add the output layer
kerasmodel.add(Dense(1, activation = 'sigmoid'))
```

# Compile the Artificial Neural Model

```python
kerasmodel.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

# Train the  Artificial Neural Model

```python
kerasmodel.fit(X_train, y_train, batch_size = 10, epochs = 150)
```

# Best Train Accuracy

```python
_, result = kerasmodel.evaluate(X_train, y_train)
print('Accuracy is:', (result * 100))
```

# Test Accuracy

```python
from sklearn.metrics import accuracy_score
pred_y = kerasmodel.predict(X_test)
print(pred_y)
```

```python
pred_y = (pred_y > 0.9)
accuracy_score(y_test, pred_y, normalize = False)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred_y)
print(cm)
```

# K-FOLD CROSS VALIDATION

```python
# import the libraries for k-fold cross validation
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define the function for k-fold cross validation
def build_classifier():
    # initialize the sequential model
    classifier = Sequential()
    # Add the first input and hidden layer
    classifier.add(Dense(12, input_dim = 8, activation = 'relu'))
    # Add the second hidden layer
    classifier.add(Dense(8, activation = 'relu'))
    # Add the output layer
    classifier.add(Dense(1, activation = 'sigmoid'))
    # compile the ANN Model
    classifier.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
# Train the ANN Model
classifier = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 100)
# depict the accuracy of k-folds
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10, n_jobs = -1)
```

```python
# mean accuracies
accuracies.mean()
```

```python
# standard deviation accuracies
accuracies.std()
```

# TO TRAIN THE ARTIFICIAL NEURAL NETWORKS USING HYPERPARAMETERS TUNING

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
# define the function to build the classifier model
def build_classifier(optimizer = 'adam'):
    # define the sequential model
    classifier = Sequential()
    # define the first input and hidden layer
    classifier.add(Dense(12, input_dim = 8, activation = 'relu'))
    # Add the second hidden layer
    classifier.add(Dense(8, activation = 'relu'))
    # Add the output layer
    classifier.add(Dense(1, activation = 'sigmoid'))
    # compile the ANN Model
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
# perform the hyperparameter tuning operation
classifier = KerasClassifier(build_fn = build_classifier)
# list the range of parameters from which we need to chjoose the optimal parameters
parameters_list = {'batch_size':[30,80], 'epochs':[20,100], 'optimizer' : ['adam','rmsprop']}
# call the gridserchoperations to select the optimal hyperparameters
gridsearch = GridSearchCV(estimator = classifier, param_grid = parameters_list, scoring = 'accuracy', cv = 10)

# using the gridsearchoperation train the model
gridsearch = gridsearch.fit(X_train, y_train)

# using the gridsearchoperation get the best parameters
parameters = gridsearch.best_params_
# using the gridsearchoperation get the best accuracy score
score = gridsearch.best_score_
```

```python
# Get the best parameters
print(parameters)
```

```python
# Get the accuracy score
print('Accuracy score is:', (score*100.0))
```
