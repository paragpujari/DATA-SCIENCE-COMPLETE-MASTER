# Perform the Data Preprocessing of the text

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re

# create the object for Porter Stemmer
ps = PorterStemmer()
```

```python
# define the empty corpus
corpus = []
for i in range(0,len(dataset)):
    # replace all the special characters except upper and lower case of the sentences with blanks
    review = re.sub('[^a-zA-Z]',' ', str(dataset['text'][i]))
    # convert the text into lower case
    review = review.lower()
    # split the text into spaces
    review = review.split()
    
    # perform the stemming on every non stop words
    review = [ps.stem(word)  for word in review if word not in stopwords.words("english")]
    # join all the words
    review = ' '.join(review)
    # append all the review words into the corpus
    corpus.append(review)
```

```python
print(corpus)
```

# COUNT VECTORIZER

```python
# Apply the Count Vectorizer on the corpus text and create the bag of words
from sklearn.feature_extraction.text import CountVectorizer

# create the object for the Count Vectorizer
cv = CountVectorizer(max_features = 500, ngram_range = (1,3))

# transform the corpus and convert it into the array
X = cv.fit_transform(corpus).toarray()
```

# PLOTTING THE CONFUSION MATRIX

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    See full source and example: 
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

# MULTINOMIAL NB ALGORITHM

```python
from sklearn.naive_bayes import MultinomialNB

classifier = MultinomialNB()

# train the model
classifier.fit(X_train, Y_train)

# predict the model
pred = classifier.predict(X_test)

# find the accuracy score between the actual and predictec value
from sklearn.metrics import accuracy_score

score = accuracy_score(pred, Y_test)
print(score * 100.0)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(pred, Y_test)
print(cm)

import itertools

# plot the confusion matrix of the graph
plot_confusion_matrix(cm , classes = ['FAKE','REAL'])
```

# Passive Aggressive Classifier Algorithm

```python
from sklearn.linear_model import PassiveAggressiveClassifier

linear_classifier = PassiveAggressiveClassifier()

# train the model
linear_classifier.fit(X_train, Y_train)

# predict the model
pred = linear_classifier.predict(X_test)

# find the accuracy score between the actual and predictec value
from sklearn.metrics import accuracy_score

score = accuracy_score(pred, Y_test)
print(score * 100.0)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(pred, Y_test)
print(cm)

import itertools

# plot the confusion matrix of the graph
plot_confusion_matrix(cm , classes = ['FAKE','REAL'])
```

# Multinomial Classifier with Hyperparameter

```python
classifier = MultinomialNB(alpha = 0.1)
```

```python
previous_score=0
for alpha in np.arange(0,1,0.1):
    sub_classifier=MultinomialNB(alpha=alpha)
    sub_classifier.fit(X_train,Y_train)
    y_pred=sub_classifier.predict(X_test)
    score = accuracy_score(Y_test, y_pred)
    if score>previous_score:
        classifier=sub_classifier
    print("Alpha: {}, Score : {}".format(alpha,score))
```
