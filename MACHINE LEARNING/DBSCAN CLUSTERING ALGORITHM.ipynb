```python
# import all the necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.neighbors import NearestNeighbors
```

```python
# load the dataset
data = pd.read_csv('Mall_Customers_dataset.csv')
```

```python
# print the data
print(data)
```

```python
# display the first five rows of the dataset
print(data.head())
```

```python
# display the last five rows of the dataset
print(data.tail())
```

```python
# shape of the dataset
print(data.shape)
```

```python
# To check if there are anu NULL Values in the dataset
print(data.isnull().sum())
```

# So from the above observations, it is clear that there are no NULL Values in the dataset

```python
# To determine the independent variables
X = data.iloc[:,3:].values
print(X)
```

```python
# To plot the input data
plt.scatter(X[:,0], X[:,1])
```

# To find the good value for epsilon

```python
# call the object for nearest neighbors
neigh = NearestNeighbors(n_neighbors = 3)
# fit and train the model using nearest neighbors
nbrs = neigh.fit(X)
# find the dsitances and indices using the nbrs data
distances , indices = nbrs.kneighbors(X)
```

```python
print(distances)
```

```python
print(indices)
```

```python
# sort the distances 
distances = np.sort(distances, axis = 0)
print(distances)
```

```python
plt.plot(distances)
```

```python
# So from the above observation, it is clear that the maximum data points lie within the epsilon value of 10
```

# DBSCAN ALGORITHM

```python
# Call the DBSCAN Algorithm
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 15.0, min_samples = 4)
# train the model using dbscan
dbscan.fit(X)
```

```python
# predict the labels for the clusters
results = dbscan.labels_
print(results)
```

```python
# identify all the regions with the core data points

# initialize all the regions with the False
results = np.zeros_like(results, dtype = 'bool')
print(results)
```

```python
# find out the region where the clusters can be formed

results[dbscan.core_sample_indices_] = True
print(results)
```

```python
# find out the efficiency of the clustering algorithm
from sklearn.metrics import silhouette_score
accuracy = silhouette_score(X, results)*100.0
accuracy
```
