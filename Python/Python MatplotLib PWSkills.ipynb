# DATA VISUALTIZATION

Data Visualization refers to the process of representing the data in pictorial and in graphical format.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# defining the dataset
x = np.arange(0,10,0.1)
y = (2 * x) + 5

# plot the graph
plt.plot(x,y)
```

# CUSTOMIZE THE LINE PLOT

```python
x = np.arange(0,10,0.1)
y = (2*x) + 5

# changing the size of the figure
plt.figure(figsize = (10,5))
# plot the graph
plt.plot(x,y, alpha = 0.5, marker = 'o', color = 'red', linestyle = ':', linewidth = 2.0)

# title of the graph
plt.title('X vs Y axis')
# x-axis
plt.xlabel('X Axis')
# y-axis
plt.ylabel('Y Axis')
# grid
plt.grid(True)

# legend of the graph
plt.legend(['Line1'], loc = 'best')

# display the graph
plt.show()
```

# TWO DIFFERENT PLOTS IN ONE FIGURE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10,1)
y1 = 2*x + 5
y2 = 3*x + 10

plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title('Graph2')
```

```python
x = np.arange(0,20,1)
y1 = 2*x + 6
y2 = 3*x + 12

plt.subplot(5,1,1)
plt.plot(x,y1)
plt.title('Graph11')

plt.subplot(5,1,2)
plt.plot(x,y2)
plt.title('Graph12')
```

```python
# To plot two or more plots in one figure
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.title('Graph4')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Graph5')
```

# PLOTS 

It is a figure that is used to draw and represents the data in the form of rectangular figures.

# 1. BARPLOT

```python
data = {'apples':20, 'mangoes':15, 'lemon':30, 'oranges':10}
a = list(data.keys())
b = list(data.values())

# To plot bar graph between the fruits and its numbers
plt.bar(a,b)
```

# 2. CUSTOMIZED BAR PLOT

```python
dict = {'apples':20, 'mangoes':15, 'lemon':30, 'oranges':10}

# get the fruits name
fruits_name = list(dict.keys())
print(fruits_name)

# get the fruits quantity
fruits_quantity = list(dict.values())
print(fruits_quantity)

# To perform the bar plot between fruits and its quantity
plt.bar(fruits_name, fruits_quantity, color = 'red')

# title of the graph
plt.title('Comparision between fruits vs quantity')

# x-axis
plt.xlabel('fruits')

#y- axis
plt.ylabel('quantity')

# display the graph
plt.show()
```

# 3. HORIZONTAL BAR PLOT

```python
dict = {'apples':20, 'mangoes':15, 'lemon':30, 'oranges':10}

fruits = list(dict.keys())
print(fruits)

quantity = list(dict.values())
print(quantity)

# To perform the horizontal bar plot between the fruits and quantity
plt.barh(fruits, quantity, color = 'pink')

# title of the graph
plt.title('Comparision between fruits vs quantity')

# x-axis
plt.xlabel('fruits')

#y- axis
plt.ylabel('quantity')

# display the graph
plt.show()
```

# 4. SCATTER PLOT

```python
# creating the dataset
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

# To perform the scatter plot between the data points
plt.scatter(a,b)
plt.scatter(a,x)
```

# 5. CUSTOMIZE THE SCATTER PLOT

```python
# creating the dataset
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]


# To perform the scatter plot between the data points
plt.scatter(a,b, color = 'green', edgecolor = 'yellow', alpha = 0.5, marker = 'o')
plt.scatter(a,x, color = 'red',   edgecolor = 'orange', alpha = 0.5, marker = '8')

# title of the graph
plt.title('Scatter Plot Graph')

# x-axis
plt.xlabel('X-Axis')

#y- axis
plt.ylabel('Y-Axis')

# To generate the legend for the graph
plt.legend(['b','x'], loc = 'best')

# To generate the grid for the graph
plt.grid(True)

# display the graph
plt.show()
```

# 8. BOX PLOT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# create three lists of data
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]

lists = list([total, order, discount])

# To perform the boxplot for the above dataset
plt.boxplot(lists, showmeans = True)

plt.grid(True)
```

# 10. PIE CHART

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# creating the dataset
label = ['Dog', 'Cat', 'Wolf', 'Lion']
size = [50, 45, 60, 80]

# To perform the pie chart between size and label
plt.pie(size, labels = label)

# To perform the pie chart title for the dataset
plt.title('Pie Chert Plot')

plt.show()
```

# 12. DOUGHNUT CHART

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# To prepare the donoughnut chart
group_names = ["Apples","Bananas","Cherry"]
group_sizes = [30,50,70]
size_centre = [5]

# To prepare the first pie chart
plt.pie(group_sizes, labels = group_names, radius = 1.5)

# To prepare the second pie chart
plt.pie(size_centre, colors = 'white', radius = 1.0)

# To perform the pie chart title for the dataset
plt.title('Doughnut Chart Plot')

plt.show()
```

# 13. AREA CHART

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# preparing the dataset
x = range(1,15)
y = [1,4,6,8,4,5,3,2,4,1,5,6,8,7]

# To plot the area chart
plt.stackplot(x,y, color = 'red', alpha = 0.5)

# To plot the line chart with the area chart
plt.plot(x,y, color = 'g')

# Grid of the graph
plt.grid(True)

# display the graph
plt.show()
```
