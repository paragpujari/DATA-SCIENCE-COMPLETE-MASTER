# Data Analysis and Visualizations

```python
# Amongst all the hotel bookings, get the count of all the hotel bookings that have been canceled and also that have not been canceled and even visually display it in bar graph

canceled_percentage = df['is_canceled'].value_counts(normalize = True)
print(canceled_percentage)

# Visually display it in a graph

# size of the graph
plt.figure(figsize = (5,4))
# title of the graph
plt.title('Reservation Count Status')
# display the data in bar graph
plt.bar(['Not Canceled','Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
# display the graph
plt.show()
```

# Observation:

From the above observation, 
it is clear that around 37% of the hotel bookings have been canceled which is much more than the half of the hotel bookings.

```python
# Get the Reservation Status is each hotels

plt.figure(figsize = (8,4))
# To depict the cancelation vs non cancelation ratio in each hotels
ax1 = sns.countplot(data = df, x = 'hotel', hue = 'is_canceled', palette = 'Blues')

# To set the legends of the graph
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))

# set the title of the graph
plt.title('Reservation Status in each hotels', size = 20)
# set the x-axis of the graph
plt.xlabel('hotel')
# set the y-axis of the graph
plt.ylabel('number of reservations')
# set the legend names of the graph
plt.legend(['not canceled','canceled'])
# display the graph
plt.show()
```

# Observation:

1. The Cancelation vs Non Cancelation Ratio for the Resort Hotel is much higher than the City Hotel.

2. The Cancelation Rate for the City Hotel is much higher than the Resort Hotel because the Resort Hotel is much expensive 
than the City Hotel.

```python
# To filter out all the resort hotels
resort_hotel = df[df['hotel'] == 'Resort Hotel']
# get the count of all the resort hotels canceled
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# To filter out all the city records
City_hotel = df[df['hotel'] == 'City Hotel']
# get the count of all the city hotels canceled
City_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Visuzally plot the cancelattion status of both the hotels in terms of adr
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_hotel   = City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# To get Average Daily Rate in City and Resort Hotel(adr for both the hotels visually)
plt.figure(figsize = (20,8))
# To set the title of the graph
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
# To plot the adr value for the Resort hotel
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
# To plot the adr value for the City hotel
plt.plot(City_hotel.index, City_hotel['adr'], label = 'City Hotel')
# To depict the legend of the graph
plt.legend(fontsize = 20)
# display the graph
plt.show()
```

# Observation:

From the above Observation, it is clear that
1. Data is in consistent form from Jan 2016 till Jan 2017.

2. In the WeekEnds, the spikes become very high means the price for Resort and the City Hotels become very high.

3. The price for the Resort Hotel has become very expensive than the City Hotel.

# To depict the Reservation Status per Month

```python
# To extract the month from Reservation status date column
df['month'] = df['reservation_status_date'].dt.month

# depict the size of the graph
plt.figure(figsize = (16,8))
# To perform the countplot operation based on 'is_canceled' column for every month
ax1 = sns.countplot(data = df, hue = 'is_canceled', x = 'month')

# set the legends of the graphs
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))

# set the title of the graph
plt.title('Reservation Status per month')

# set the x-axis of the graph
plt.xlabel('month')

# set the y-axis of the graph
plt.ylabel('number of reservations')

# name the legends of the graph
plt.legend(['not canceled','canceled'])
# display the graph
plt.show()
```

# Observations:

From the above observation, it is clear that

1. Maximum number of cancellations has taken place in the month of January and at that month our reservations is the minimum.

2. Maximum number of reservations has taken place in the month of August and at that month our cancellations is the minimum.

# Visually Represent the ADR Value Month wise for the Canceled hotels

```python
# to depict the size of the graph
plt.figure(figsize = (18,6))

# to set the title of the graph
plt.title('ADR per month', fontsize = 30)
# To perform the visual representation of adr month wise to chekc if price is having any affect on the cancellation of the hotel rooms.
sns.barplot('month','adr',data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

# to display the graph
plt.show()
```

# Observation:

From the above observation, it is clear that
1. The ADR(Average Daily Rate)or the Price Value is maximum in the month of January, so maximum cancellations and minimum reservations  has taken place in the month of January.

2. The ADR(Average Daily Rate) or the Price Value is minimum in the month of August, so minimum cancellations and maximum reservations has taken place in the month of August.

# To compare the ADR(Average Daily Rate) between the cancelled hotel status vs non canceled hotel status

```python
canceled_rooms.head(1)

# Get the mean values of the adr for the canceled rooms status
canceled_rooms_adr = canceled_rooms.groupby('reservation_status_date')[['adr']].mean()
# reset the index for the adr of the canceled rooms
canceled_rooms_adr.reset_index(inplace = True)
# sort the values of the canceled rooms adr according to their reservation dates
canceled_rooms_adr.sort_values('reservation_status_date', inplace = True)

# get the mean values of the adr for the non canceled rooms status
non_canceled_rooms = df[df['is_canceled'] == 0]
# get the adr for the non canceled rooms status
non_canceled_rooms_adr = non_canceled_rooms.groupby('reservation_status_date')[['adr']].mean()
# reset the index for the adr of the non canceled rooms
non_canceled_rooms_adr.reset_index(inplace = True)
# sort the values of the non canceled rooms adr according to their reservation dates
non_canceled_rooms_adr.sort_values('reservation_status_date', inplace = True)
```

```python
non_canceled_rooms_adr
```

```python
canceled_rooms_adr
```

```python
# plot the adr value for the canceled hotel rooms vs non canceled hotel rooms
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(non_canceled_rooms_adr['reservation_status_date'], non_canceled_rooms_adr['adr'], label = 'not canceled')
plt.plot(canceled_rooms_adr['reservation_status_date'], canceled_rooms_adr['adr'], label = 'canceled')
plt.legend()
```

# Observation
Data is consistent between 2016 till 2017.

# To filter out all the adrs of the canceled rooms and the non cancelled berween 2016 till september 2017

```python
non_canceled_df_adr = non_canceled_rooms_adr[(non_canceled_rooms_adr['reservation_status_date'] > '2016') & (non_canceled_rooms_adr['reservation_status_date'] < '2017-09')]

canceled_df_adr = canceled_rooms_adr[(canceled_rooms_adr['reservation_status_date'] > '2016') & (canceled_rooms_adr['reservation_status_date'] < '2017-09')]
```

```python
print(non_canceled_df_adr)
```

```python
# To plot the adr for the non canceled vs canceled hotel rooms

plt.figure(figsize = (14,8))

# set the title of the graph
plt.title('Average Daily Rate', fontsize = 30)

# To plot the adr for the non canceled rooms for its reservation_status_date
plt.plot(non_canceled_df_adr['reservation_status_date'], non_canceled_df_adr['adr'], label = 'not canceled')

# To plot the adr for the canceled hotel rooms for its reservation status date
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label = 'canceled')

# set the legend of the graph
plt.legend(fontsize = 20)

# display the graph
plt.show()
```
