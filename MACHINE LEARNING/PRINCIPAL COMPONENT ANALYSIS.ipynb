```python
# import all the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load the dataset
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
print(cancer)
```

```python
# depict what kind of data is present in cancer
print(cancer.keys())
```

```python
# to depict the data present in 'DESCR'
print(cancer['DESCR'])
```

```python
# Creating a Dataframe
df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
print(df)
```

```python
# display the first five rows of the dataframe
print(df.head())
```

```python
# display the last five rows of the dataframe
print(df.tail())
```

```python
# shape of the dataframe
print(df.shape)
```

```python
# Ti check if there are any NULL Vlaue sin the dataframe
print(df.isnull().sum())
```

```python
# Before performing PCA, we need to perform feature scaling to make all the inputs in one range
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

scaler = sc.fit(df)
```

```python
scaler = sc.transform(df)
```

```python
# import the library for PCA
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
```

```python
# To fit and train the model
scaled = pca.fit(scaler)
```

```python
# To transform the PCA Data Model so as to reduce it's dimensions
scaled = pca.transform(scaler)
```

```python
# shape of the original data
print(scaler.shape)
```

```python
# shape of the transformed data
print(scaled.shape)
```

```python
scaled
```

```python
# To plot the principal component analysis graph
plt.figure(figsize = (8,6))
plt.scatter(scaled[:,0], scaled[:,1], c = cancer['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```
