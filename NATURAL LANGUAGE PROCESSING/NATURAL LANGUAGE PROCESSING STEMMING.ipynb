# Stemming 

It is the process of reducing the words to their root words

```python
# import the library for stemming

from nltk.stem import PorterStemmer

# call the object for Porter Stemmer
ps = PorterStemmer()
```

```python
# list of words for stemming
words = ["like", "likely", "liking", "unlike","liked"]
```

```python
# stem all the words to the root words
for i in words:
    print(ps.stem(i))
```

# Gutenberg Corpus

```python
from nltk.corpus import gutenberg
```

```python
# import the raw file in the gutenberg corpus
sample = gutenberg.raw('carroll-alice.txt')
print(sample)
```

```python
# perform the word tokenization on the above text
from nltk.tokenize import word_tokenize
words = word_tokenize(sample)
print(words)
```

```python
from nltk.corpus import stopwords

english_stopwords = set(stopwords.words("english"))

print(english_stopwords)
```

```python
# Remove all the stop words from the text

filtered_words = []

for i in words:
    if(i not in english_stopwords):
        if(len(i) > 3):
            filtered_words.append(i)
            
print(filtered_words)
```

```python
# import the library for porter stemmer
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
# stem all the non stop words obtained from the text to their root words
for i in filtered_words:
    print(ps.stem(i))
```
