```python
# subset of the input
datasubsets = data[['Age','Annual Income (k$)']]
print(datasubsets)
```

```python
# performing the scatter plotting on the datasubsets
plt.figure(figsize = (10,5))
plt.scatter(data[['Age']], data[['Annual Income (k$)']], s = 100, c = 'blue')
```

```python
# To check whether hierarchical clusters can be drawn from the above raw data, we need to plot the dendogram
import scipy.cluster.hierarchy as sch

# To develop the dendrogram to check whether clusters can be formed or not
dend = sch.dendrogram(sch.linkage(datasubsets, method = 'ward'))

# To give the title of the graph
plt.title('Dendrogram')

# x-axis of the grapph
plt.xlabel('Customers')

#y-axis of the graph
plt.ylabel('Euclidean Distance')

# display the graph
plt.show()


# check for the longest line that is passing horizontally without crossing vertically.
```

# HIERARCHICAL CLUSTERING

```python
# To perform Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')

# train and predict the cluster
agg.fit_predict(datasubsets)
```
